```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("sales_data.csv")

df["Date"]=pd.to_datetime(df["Date"])
```

```python
print("Data Info:")
print(df.info)
```

```python
total_sales_np = np.array(df["Total Sales"])

quantity_sold_np = np.array(df["Quantity Sold"])

# print(total_sales_np)

# print(quantity_sold_np)
```

```python
print("Statistical Analysis: ")
print(f"Total Sales - Mean: ${np.mean(total_sales_np):.2f}, Median: ${np.median(total_sales_np)}")
```

```python
sales_by_category = df.groupby("Category")["Total Sales"].sum()
print(sales_by_category)
```

```python
plt.figure(figsize=(8,5))
sales_by_category.plot(kind='bar', color=['skyblue','red'])
plt.title("Total Sales by Category")
plt.xlabel("Category")
plt.ylabel("Total Sales ($)")

plt.xticks(rotation=10)
plt.yticks(rotation=60)

plt.show
```

```python
plt.figure(figsize=(10,5))
sns.lineplot(x="Date",y="Total Sales", data=df, marker="o", color="red")

plt.title("Sales Trend Over Time")
plt.xlabel("Date")
plt.ylabel("Total Sales ($)")

# plt.xticks(rotation=10)
# plt.yticks(rotation=60)
plt.grid()

plt.show
```
